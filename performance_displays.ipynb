```python
"""
    Template for plotting the performance of AI
"""

from matplotlib import pyplot as plt
```

```python
my_data: dict[ str, list ] = {
    "x": [ 1,2,3,4,5,6 ],
    "mmr": [ 900, 1000, 1400, 1500, 1800, 1900 ]
}

plt.plot(
    my_data["x"],
    my_data["mmr"]
)
plt.xlabel( "layer count" )
plt.ylabel( "mmr" )
plt.title( "MMR By Layer, 122 Neurons Per Layer")
```
